```python
# handling all empty text
print(len(train_sample))

empty_mask = train_sample['text'].str.strip() == ''
print(f"\nEmpty/whitespace rows:{empty_mask.sum()}")
print("empty rows:")
print(train_sample[empty_mask][['meeting_id','speaker_id','text']].head())

```

```python
# lowercasing the text
print("sample before lowercase")
print(train_sample["text"].head(5).tolist())

train_sample2 = train_sample.copy()
train_sample2["text"] = train_sample2["text"].str.lower()

print("sample after lowercase")
print(train_sample2["text"].head(5).tolist())

```

```python
# dropping single words utterances
backchannels = {
    'yeah','yep','yup','yes','no','nope','okay','ok','right',
    'sure','fine','hmm','mm','mhm','uh','oh','ohh','ahh','ah',
    'good','great','cool','nice','thanks','thank','alright','exactly'
}

def is_pure_backchannel(text):
  words = set(str(text).strip().split())
  return words.issubset(backchannels)

backchannel_mask = train_sample2["text"].apply(is_pure_backchannel)
print(f"pure backchannel rows:{backchannel_mask.sum()}")
print("backchannel rows being dropped:")
print(train_sample2[backchannel_mask]['text'].value_counts().head(10))

train_sample3 = train_sample2[~backchannel_mask].copy()

print(f"rows dropped:{len(train_sample2)-len(train_sample3)}")
```

```python
train_sample3.head(10)
```

```python
# removing filler words from text
import re
fillers = r'\b(um|uh|er|ah|hmm|mm|hm|uhh|umm|eh|mhm|huh|oh)\b'

has_fillers = train_sample3['text'].str.contains(fillers, flags=re.IGNORECASE, regex=True)
print(f"Rows containing fillers : {has_fillers.sum()}")
print(f"\nSample rows WITH fillers (before):")
print(train_sample3[has_fillers]['text'].head(5).tolist())

# apply the removal
train_sample4 = train_sample3.copy()
train_sample4['text']=train_sample4['text'].str.replace(fillers,'',regex=True,case=False)


# Fix extra spaces
train_sample4['text'] = train_sample4['text'].str.replace(r'\s+', ' ', regex=True).str.strip()

print(f"\nSame rows AFTER filler removal:")
print(train_sample4[has_fillers]['text'].head(5).tolist())

```

```python
FRAGMENT_PATTERN = r'\b\w+-(?=\s|$)'

# See fragments before
has_fragments = train_sample4['text'].str.contains(FRAGMENT_PATTERN, regex=True)
print(f"Rows with broken fragments : {has_fragments.sum()}")
print(f"\nSample fragments (before):")
print(train_sample4[has_fragments]['text'].head(5).tolist())
```

```python
import re
import pandas as pd


FILLERS = r'\b(um|uh|er|ah|hmm|mm|hm|uhh|umm|eh|mhm|huh|oh)\b'

BACKCHANNELS = {
    'yeah', 'yep', 'yup', 'yes', 'no', 'nope',
    'okay', 'ok', 'right', 'sure', 'fine',
    'hmm', 'mm', 'mhm', 'uh', 'oh', 'ah',
    'good', 'great', 'cool', 'nice',
    'thanks', 'thank', 'alright'
}

FRAGMENT_PATTERN = r'\b\w+-(?=\s|$)'
MIN_WORDS        = 3

def clean_ami_dataframe(df):

    # track rows at each step
    # stats = {'raw': len(df)}

    # Drop NULL text
    df = df.dropna(subset=['text']).copy()
    # stats['after_null_drop'] = len(df)
    # print(f"\n NULL drop        : {stats['raw']} → {stats['after_null_drop']}  (-{stats['raw'] - stats['after_null_drop']} rows)")

    # Drop empty/whitespace
    mask = df['text'].str.strip() == ''
    df   = df[~mask].copy()
    # stats['after_empty_drop'] = len(df)
    # print(f"Empty drop       : {stats['after_null_drop']} → {stats['after_empty_drop']}  (-{stats['after_null_drop'] - stats['after_empty_drop']} rows)")

    # Lowercase
    df['text'] = df['text'].str.lower()
    # stats['after_lower'] = len(df)
    # print(f"Lowercase        : {stats['after_empty_drop']} → {stats['after_lower']}  (-0 rows)")

    #  Drop pure backchannels
    def is_backchannel(text):
        return set(str(text).strip().split()).issubset(BACKCHANNELS)

    mask = df['text'].apply(is_backchannel)
    df   = df[~mask].copy()
    # stats['after_backchannel'] = len(df)
    # print(f"Backchannel drop : {stats['after_lower']} → {stats['after_backchannel']}  (-{stats['after_lower'] - stats['after_backchannel']} rows)")

    # Remove filler words
    df['text'] = df['text'].str.replace(FILLERS, ' ', regex=True, case=False)
    df['text'] = df['text'].str.replace(r'\s+', ' ', regex=True).str.strip()
    # stats['after_fillers'] = len(df)
    # print(f"Filler removal   : {stats['after_backchannel']} → {stats['after_fillers']}  (-0 rows, text cleaned)")


    #Drop short utterances
    df['word_count'] = df['text'].str.split().str.len()
    mask = df['word_count'] < MIN_WORDS
    df   = df[~mask].drop(columns=['word_count']).copy()
    # stats['after_short_drop'] = len(df)
    # print(f"Short drop       : {stats['after_fragments']} → {stats['after_short_drop']}  (-{stats['after_fragments'] - stats['after_short_drop']} rows)")

    # FINAL SUMMARY

    # print(f"  TOTAL REMOVED : {stats['raw'] - stats['after_short_drop']} rows")
    # print(f"  DATA RETAINED : {stats['after_short_drop'] / stats['raw'] * 100:.1f}%")


    return df
```

```python
df_clean_sample = clean_ami_dataframe(train_sample)

print(df_clean_sample[['meeting_id', 'speaker_id', 'text', 'begin_time']].head(10))
```
